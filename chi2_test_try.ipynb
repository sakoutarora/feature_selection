Chi squared test is same as chi squared test for independcy in categorical tables as studies in MC205

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
df.head ()
```

```python
cols = ['pclass','sex','parch','embarked','class','alive','alone','survived']
```

```python
df[cols]
```

```python
df['sex'] = np.where(df['sex'] == 'male' ,1,0 )
```

```python
df['alone'] = np.where(df['alone'] == True,1,0 )
```

```python
df['alive'] = np.where(df['alive'] == 'yes',1,0)
```

```python
label = { k:i   for i,k in enumerate(df['embarked'].unique(),0)}
df['embarked'] = df['embarked'].map(label)
```

```python
df['class'].unique()
label = {'First':0,'Second':1,'Third':2}
df['class'] = df['class'].map(label)
```

```python
from sklearn.model_selection import train_test_split as TTS
xtrain,xtest,ytrain,ytest = TTS(df[['pclass','sex','parch','embarked','class','alive','alone']],df['survived'],test_size=0.3,random_state= 79)         
```

```python
xtrain
```

```python
from sklearn.feature_selection import chi2
```

```python
fp_val = chi2(xtrain,ytrain)
```

```python
pval = pd.Series(fp_val[1])
pval.index = ['pclass','sex','parch','embarked','class','alive','alone']
pval = pval.sort_values(ascending=True)
```
